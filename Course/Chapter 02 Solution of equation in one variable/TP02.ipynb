```python
# a.) use secant method
def Secant(f,p0,p1,TOL=1e-16,N=100):
    stop = N+1 

    for i in range(stop):
        q0 = f(p0)
        q1 = f(p1)
        p = p1-q1*(p1-p0)/(q1-q0)
        
        if abs(p-p1) < TOL:
            return p 
        p0 = p1 
        p1 = p         
    return None

if __name__ == '__main__':
    def f(x):
        return x**2 - 6

    p3 = Secant(f=f,p0=3,p1=2,TOL=1e-16)
    print(f'p3 = {p3}')
```

```python
# b.) use false position method
def false_position(f, p0, p1, tol, max_iter):
    q0 = f(p0)
    q1 = f(p1)
    
    for i in range(1, max_iter + 1):
        p = p1 - q1 * (p1 - p0) / (q1 - q0)
        
        if abs(p - p1) <= tol:
            return p
        
        q = f(p)
        if q * q1 < 0:
            p0 = p1
            q0 = q1
            
        p1 = p
        q1 = q

# Example usage
def f(x):
    return x**2 - 6

p0 = 3
p1 = 2
tol = 1e-6
max_iter = 100

root = false_position(f, p0, p1, tol, max_iter)
print(f"p3 = {root}")
```

$ Since\space we\space know\space that\space A = \frac{P}{i}[(1+i)^n - 1] \space so,\space substitute\space value\space into\space equation\space we\space have: $
$$ A = \frac{P}{i}[(1+i)^n - 1] $$
$$ \iff 750,000 = \frac{12 * 1500}{i}[(1+i)^{20} - 1] $$
$$ \iff 750,000i = 12 * 1500[(1+i)^{20} - 1] $$
$$ \iff 750i = 18(1+i)^{20} - 1 $$
$$ \iff 18(1+i)^{20} - 750i-1=0 $$
$ Find\space the\space value\space of\space minimum(i) $

```python
def f(x):
    return 18*(1+x)**(12*20) - 750*x - 1

def f_prime(x):
    return 18*20*(1+x)**() - 750

def newton_raphson_method(f, f_prime, x0, tol=1e-8, max_iter=100):
    x = x0
    i = 0
    while i < max_iter:
        delta_x = f(x) / f_prime(x)
        x = x - delta_x
        if abs(delta_x) < tol:
            return x, i  # Return the root and the number of iterations
        i += 1

    print("The Newton-Raphson method did not converge within the maximum number of iterations.")
    return None, i  # Return None for the root and the number of iterations

# Find the root of f(x) = 0 (or closest to zero) and the minimal i using the Newton-Raphson method
x0 = 0.5  # Initial guess
root, iterations = newton_raphson_method(f, f_prime, x0)
if root is not None:
    print("The value of x that makes f(x) closest to zero is approximately:", root)
    print("Minimal i required:", iterations)
```

```python
def summarizeEvaluate(a,x,d):

    N = len(a)
    n = N-1 
    v = [0] * (d+1)
    v[0] = a[n]
    
    for k in range(1,N,1):
        for j in range(d,0,-1):
            v[j] = j*v[j-1] + x *v[j]
        v[0] = a[n-k] + x*v[0]
    
    return v 

if __name__ == "__main__":
    
    a = [1,-20,200,-550,600]
    x = 1.1/2
    d = 5
    val3 = summarizeEvaluate(a=a,x=x,d=d)  
    
```

```python
def Muller(a,p0,p1,p2,ftol=1.0e-10, maxiter=100):
    
    fp0 = summarizeEvaluate(a=a,x=p0,d=0)[0]
    fp1 = summarizeEvaluate(a=a,x=p1,d=0)[0]
    fp2 = summarizeEvaluate(a=a,x=p2,d=0)[0]
    
    h1 = p1 - p0
    h2 = p2 - p1 
    d1 = (fp1-fp0)/h1 
    d2 = (fp2-fp1)/h2 
    
    d = (d2-d1)/(h1+h2)
    df = pd.DataFrame(data={'p':[p0,p1,p2],'f(p)':[fp0,fp1,fp2]})
    i = 3
    
    condition = True 
    while condition:
        b = d2 + h2 * d
        fp2 = summarizeEvaluate(a=a,x=p2,d=0)[0]
        D = (b**2 - 4*fp2*d)**(1/2)
        
        if abs(b-D) < abs(b+D):
            E = b + D
        else:
            E = b - D 
            
        h = -2*fp2/E 
        p = p2 + h 
        fp = summarizeEvaluate(a=a,x=p,d=0)[0]
        df.loc[i,:] = {'p':p,'f(p)':fp}
        print(f'i={i:0f}, p = {p:.10f}, f(p) = {fp:.10f}')
        
        condition = abs(h) > ftol 
        
        if condition:
            p0 = p1 
            p1 = p2 
            p2 = p 
            
            fp0 = summarizeEvaluate(a,x=p0,d=0)[0]  # [0] mean that we want to access to the first element.
            fp1 = summarizeEvaluate(a,x=p1,d=0)[0]
            fp2 = summarizeEvaluate(a,x=p2,d=0)[0]
            
            h1 = p1 - p0 
            h2 = p2 - p1 
            d1 = (fp1 - fp0)/h1 
            d2 = (fp2 - fp1)/h2 
            d = (d2 - d1)/(h2-h1)
            i = i + 1
            
    if i>=maxiter:
        print(f'method failed after {maxiter} iteration')
    else:
        print(f'\nRoot = {p:.2f}')
    return (p,df)


if __name__ == "__main__":
    a = [1,-20,200,-550,600]
    
    p,df = Muller(a=a,p0=0.1,p1=1,p2=1.1/2)
    
    print(f'p = {p:.10f}')
    print(df)   
```

```python
# Laguerre method 01 :
def Laguerre(a,x,e=1e-10,maxiter = 100):
    n = len(a) - 1 
    stop = maxiter + 1 
    p,dp,dpp = summarizeEvaluate(a=a,x=x,d=2)
    df = pd.DataFrame(data={'x':[x], 'p(x)':[p]})
    
    for k in range(1,stop,1):
        if abs(p) < e:
            return (x,df)
        G = dp/p 
        H = G**2 - dpp/p 
        F = ((n-1)*(n*H-G**2)) ** (1/2)
        
        if abs(G+F) > abs(G-F):
            dx = n/(G+F)
        else:
            dx = n/(G-F)
        
        x = x - dx
        p,dp,dpp = summarizeEvaluate(a=a,x=x,d=2)
        df.loc[k] = {'x':x,'p(x)':p}
        
        if abs(dx) < e:
            return (x,df)
       
    return (x,df)

if __name__ == "__main__":
    a = [-136,-85,-9,5,1]
    x = 1
    e = 1e-10
    x,df = Laguerre(a=a,x=x,e=e)
    print(df)
    print(f'Required root is : {(x)}')
```
