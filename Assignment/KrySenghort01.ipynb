```python
import math as mt
def verify(a,b):
  y1 = mt.exp(-a)-2*a-2
  y2 = mt.exp(-b)-2*b-2
  if(y1*y2<0):
    print(f'f(x) = 0 has a root in the interval[{a,b}]')
  else:
    print(f'f(x) has not a root in the interval[{a,b}]')
  return 

if __name__ == "__main__":
  verify(-1,0)        
```
