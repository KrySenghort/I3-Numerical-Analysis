### <font color = green|>3. Plot the natural cubic spline S(x).

````python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
x = [-1.00, 0.00, 1.00, 2.00]
y = [0.25, -1.25, -0.75, 1.75]

# Define the natural cubic spline function S(x)
def S(x_val):
    if -1.00 <= x_val <= 0.00:
        return -1.25 - 2.50*x_val - 2.50*x_val**2 - 1.50*x_val**3
    elif 0.00 <= x_val <= 1.00:
        return -0.75 + 0.75*x_val - 0.75*x_val**2 + 0.75*x_val**3

# Generate points for plotting
x_plot = np.linspace(-1.00, 2.00, 100)
y_plot = [S(x_val) for x_val in x_plot]

# Plot the natural cubic spline
plt.plot(x_plot, y_plot, label='S(x)')
plt.scatter(x, y, color='red', label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Natural Cubic Spline')
plt.legend()
plt.grid(True)
plt.show()
````
