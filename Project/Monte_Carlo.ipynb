## **3. Algorithm 1**  
The Monte Carlo process uses the theory of large numbers and random sampling to approximate values that are very close to the actual solution of the integral.  
In this algorithm,  


where  
* **<f(x)>** is the sumation of result values that getting from random value **x** between interval **a** and **b**  
* **N** is the number of random use to approximate the value  
* **f(x)** is a function that we are going to compute
 

**Suppose** you want to calculate the folowing integral using Monte Carlo 


```python
import numpy as np
import seaborn as sns
sns.set(style='darkgrid')

a = 0
b = np.pi
N = 1000

def f(x):
    return np.sin(x)

result = []

for j in range(N):
    x = np.zeros(N)
    for i in range(len(x)):
        x[i] = np.random.uniform(a,b)
    sumAllValue = 0.0
    for i in x :
        sumAllValue += f(i)   
    
    answer = (b-a)/float(N)*sumAllValue
    result.append(answer)


print(f'The Value of this Integral is : {answer}')
sns.histplot(data = result, kde=True).set(title='Distribution of Integral Value', xlabel='Integral Value')

```

```python

import numpy as np
import matplotlib.pyplot as plt

a = -(np.pi)/2
b = (np.pi)/2
M = 1
N = 1000

def f(x):
    return np.cos(x)

# Generate random points
x = np.random.uniform(a, b, N)
y = np.random.uniform(0, M, N)

# Compute the mask of points below the curve
mask = y <= f(x)

# Compute the estimated area
counter = np.sum(mask)
area = (M - 0) * (b - a) * counter / N

# Plot the function and the random points
fig, ax = plt.subplots()
x_vals = np.linspace(a, b, 1000)
y_vals = f(x_vals)
ax.plot(x_vals, y_vals, 'b-', label='Function')
ax.scatter(x[mask], y[mask], c='g', label='Points below curve')
ax.scatter(x[~mask], y[~mask], c='r', label='Points above curve')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Monte Carlo Integration')
ax.legend()
plt.show()

print(f'The estimated area is: {area}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    return (x**2 + y**2)

a = 0
b = 1
c = 0
d = 1
M = 1
N = 500

# Generate random points
x = np.random.uniform(a, b, N)
y = np.random.uniform(c, d, N)
z = np.random.uniform(0, M, N)

# Compute the mask of points inside the volume
mask = z <= f(x, y)

# Compute the estimated volume
counter = np.sum(mask)
volume = M * (d - c) * (b - a) * counter / N

# Plot the function and the random points
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x_vals = np.linspace(a, b, 100)
y_vals = np.linspace(c, d, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

ax.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.5)
ax.scatter(x[mask], y[mask], z[mask], c='g', marker='o', label='Points inside volume')
ax.scatter(x[~mask], y[~mask], z[~mask], c='r', marker='o', label='Points outside volume')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Monte Carlo Integration in 3D')
ax.legend()
plt.show()

print(f'The estimated volume is: {volume}')
```

## **Monte Carlo Fair Coin**

```python
import numpy as np

def monte_carlo_coin(n):
    counter = 0
    for i in range(n):
        xi = np.random.uniform(0, 1)
        if xi <= 0.5:
            counter += 1
    p_head = counter / n
    return p_head

n = 1000
p_head = monte_carlo_coin(n)
print(f'The probability of getting a head with {n} coin flips is: {p_head}')
```

```python
import numpy as np

def simulate_weather(n, p_rain):
    weather = np.random.choice([0, 1], size=n, p=[1-p_rain, p_rain])
    return weather

def count_consecutive_rain(weather):
    counter = 0
    for i in range(len(weather)-2):
        if weather[i] == 1 and weather[i+1] == 1 and weather[i+2] == 1:
            counter += 1
    return counter

# Set the simulation parameters
n_days = 1000000
p_rain = 0.3

# Simulate the weather and count the number of consecutive rainy days
weather = simulate_weather(n_days, p_rain)
consecutive_rain_count = count_consecutive_rain(weather)

# Estimate the probability of rain for three consecutive days
p_consecutive_rain = consecutive_rain_count / (n_days - 2)

# Output the results
print(f'The probability of rain for three consecutive days is: {p_consecutive_rain:.4f}')
```
