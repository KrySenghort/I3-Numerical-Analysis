To study the continuity of a function using Python, you can follow these steps:
1. Define the function: Start by defining the function you want to study. For example, let's say we want to study the continuity of the function 
$$ f(x) = x^2 $$

```python
def f(x):
    return x**2
```

2. Choose a point where you want to study the continuity of the function. For example, we can choose x=1.

```python
a = 1
```

3. Check if the function is defined at the chosen point by evaluating it using Python. If the function returns a value, it is defined at that point.

```python
f(a)
```

4. Check if the limit exists

```python
x = np.linspace(0.9, 1.1, 100)
y = f(x)
print(y)
```

5. Check if the function is continuous: By comparing its value at the point with the limit of the function as x approaches the point. If these two values are equal, then the function is continuous at that point.

```python
x = sp.Symbol('x')
y = x**2
if f(a) == sp.limit(y,x,1):
    print("The function is continuous at x=1")
else:
    print("The function is not continuous at x=1")
```

### <font color = yellow>Exercise 02: Study continuity of function below
$$ f(x) = \begin{cases}
x^2,    x<0\\
2x^2+x, 0<= x < 1\\
4,      x>=1\\
\end{cases} $$

```python
def f(x):
    if x < 0:
        return x ** 2
    elif 0 <= x < 1:
        return 2 * x ** 2 + x
    else:
        return 4

epsilon = 1e-6 # a small positive number
```

```python
# check left-hand limit at x = 0
left_limit = f(0-epsilon)
function_value = f(0)
if abs(left_limit-function_value) > epsilon:
  print('Function is not continuous at x = 0')
else:
  print("Function is continuous at x = 0")
```

```python
# check right-hand limit at x=1
right_limit = f(1 + epsilon)
function_value = f(1)
if abs(right_limit - function_value) > epsilon:
  print("Function is not continuous at x = 1")
else:  
  print("Function is continuous at x = 1")
```

To find the limit of a sequence, you need to examine the behavior of the terms as the index grows larger and larger. Depending on the pattern of the terms, the sequence may converge to a specific value, diverge to infinity, or oscillate between two or more values.

Here are some general strategies for finding the limit of a sequence:

1. Try to recognize a common pattern or formula for the terms of the sequence. For example, if the terms are given by an arithmetic or geometric progression, you can use the corresponding formulas to find the limit.

2. If the sequence is defined recursively, try to find a closed-form expression for the nth term in terms of n. Then you can take the limit of this expression as n goes to infinity.

3. If the sequence is not easily recognized by a formula, you can try to apply the squeeze theorem or the monotone convergence theorem. These theorems provide criteria for determining whether a sequence converges or diverges based on its relationship with other known sequences.

4. In some cases, you may need to use more advanced techniques such as L'Hopital's rule, Taylor series, or complex analysis to find the limit of a sequence. 

It's worth noting that not all sequences have limits. Some sequences may grow without bound (diverge to infinity) or oscillate indefinitely between different values. It's important to analyze the behavior of a sequence carefully before attempting to find its limit.

## <font color = yellow>V.) Roll's Theorem

The Roll Theorem is a theorem in calculus that helps to find the value of a limit involving a derivative. <font color = yellow>It states that if a function f(x) is continuous on the closed interval [a,b] and differentiable on the open interval (a,b),</font> then there exists at least one point c in the open interval (a,b) such that:

<font color = yellow>$$ f(b) - f(a) = f'(c) * (b-a) $$ </font>

This theorem has several applications, including:

<font color = yellow>+ Finding the average rate of change:</font> By using the Roll Theorem, we can find the average rate of change of a function over an interval. For example, if we want to find the average velocity of an object between times t=2 and t=4, we can use the Roll Theorem to find the instantaneous velocity at some point c in the interval, and then use that value to calculate the average velocity.

<font color = yellow>+ Proving the Mean Value Theorem:</font> The Roll Theorem is a special case of the Mean Value Theorem, which states that for any continuous and differentiable function f(x) on the interval [a,b], there exists at least one point c in the interval such that:

<font color = yellow>$$ f'(c) = \frac{f(b) - f(a)}{b-a} $$ </font>

By using the Roll Theorem, we can prove the Mean Value Theorem for a specific case where b=a+h, where h is some nonzero constant.

Optimizing functions: The Roll Theorem can be used to help optimize functions by finding critical points where the derivative is zero or undefined. This can be useful in applications such as finding the maximum or minimum values of a function, or finding the inflection points of a curve.
Overall, the Roll Theorem is a powerful tool in calculus that has many important applications in various fields such as physics, economics, engineering, and more.

## <font color = yellow>VI.) Intermediate Value Theorem

The Intermediate Value Theorem (IVT) is a fundamental theorem in calculus that states that for any <font color = yellow>continuous function f defined on a closed interval [a, b], if there exists some y value between f(a) and f(b), then there must exist at least one x value in the interval [a, b] such that f(x) equals y.</font>

In other words, if we imagine graphing a continuous function f over an interval [a, b], and we have a horizontal line that passes through some y-value between f(a) and f(b), then the IVT guarantees that this line intersects the graph of f at least once in the interval [a, b].

This theorem has important applications in many areas of mathematics, including analysis, topology, and geometry. It is also essential for proving many key results in calculus, such as the existence of roots of equations and the Extreme Value Theorem.

## <font color = yellow>XI.) Round-off Errors

Absolute error and relative error are two ways to measure the difference between an approximate value and an exact value.
+ <font color = yellow>Absolute error</font> is defined as the absolute value of the difference between the approximate value (denoted by A) and the exact value (denoted by E). Mathematically, it can be expressed as:
<font color =yellow>$$ Absolute \space error = |A - E| $$</font>
Absolute error tells us how far off our approximation is from the actual value in the same units as the measurement.

+ <font color = yellow>Relative error</font> is defined as the ratio of the absolute error to the exact value. Mathematically, it can be expressed as:
<font color = yellow>$$ Relative \space error = (|A - E|) / |E| $$</font>
Relative error tells us how much error there is relative to the size of the actual value. It is often expressed as a percentage.

Both absolute error and relative error are important concepts in numerical analysis and scientific computing. They help us understand the accuracy of our computations and the precision of our measurements.

<br>


#### <font color = yellow>Exercise 01: 
$$ study \space about \space value \space of \space \pi $$

```python
# Approximate value of pi
approx_pi = 22/7

# Exact value of pi
exact_pi = 3.14159265358979323846

# Absolute error
abs_error = abs(approx_pi - exact_pi)

# Relative error
rel_error = abs_error / abs(exact_pi)

print("Absolute error:", abs_error)
print("Relative error:", rel_error)
```
