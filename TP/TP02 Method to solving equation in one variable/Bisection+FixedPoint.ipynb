### <b><font color = 'blue'>Exercise 1: Bisection method</font></b>

$ Solve $ $ the $ $ equation $ $ f(x) $ $ = $ $ cos(x) $ $ - $ $ x $ $ = $ $ 0 $ $ by $ $ using $ $ bisection $ $ method $ 

```python
import pandas as pd
def bisection(f, x0, x1, e):
    step = 1
    condition = True
    df = pd.DataFrame(data={'x':[x0],'f(x)':[f(x0)]}) # 
    while condition:
        x2 = (x0 + x1) /2
        print(f'Step:{step}, x2 = {x2:8.16f} and f(x2) = {f(x2):8.16f}')
        if f(x0) * f(x2) < 0:
            x1 = x2
        else:
            x0 = x2
        df.loc[step] = {'x':x0, 'f(x0)':f(x0)}  # use it to test afterward index.
        step = step + 1
        condition = abs(f(x2)) > e
    
    print(f'\nRequired root is : {x2:0.16f}')
    return df
if __name__ == "__main__":
    import math as mt
    def f(x): return mt.cos(x)-x
    df = bisection(f=f,x0=0,x1=mt.pi/4,e=1.0e-16) 
```

```python
print(df.style.format({'x':'{:.16f}','f(x)':'{:.16f}'}).to_latex())
with pd.ExcelWriter(path = "Bisection method.xlsx",mode = 'w') as writers:
  df.to_excel(excel_writer=writers)
```

### <b><font color = 'blue'>Exercie 2: Fixed point</font></b>

```python
def FixedPoint(f,g,p0,e=1.0e-8,N=100)->pd.DataFrame:
  step = 1
  flag = 1
  condition = True
  df = pd.DataFrame(data={'x':[p0],'f(x)':[f(p0)]})
  df.index.name = "step"
  while condition:
    p1 = g(p0)
    f1 = f(p1)
    df.loc[step] = [p1,f1]
    print(f"Step:{step:3d},p1={p1:0.8f} and f(p1)={f1:0.8f}")
    p0 = p1
    step+=1
    if(step>N):
      flag = 0
      break
    condition = abs(f1)>e
  if(flag==1):
    print(f"\nRequired root is : {p1:0.16f}\n")
  else:
    print(f"\nNot convergent.\n")
  return df

if __name__=="__main__":
  import math
  def f(x):
    return math.cos(x)-x
  def g(x):
    return math.cos(x)
  p0 = 1.5
  e = 1.0e-16
  df = FixedPoint(f=f,g=g,p0=p0,e=e)
  print(df)
  df.style.format({'x':'{:.16f}','f(x)':'{:.16f}'}).to_latex()
  with pd.ExcelWriter(path = "Fixedpoint.xlsx",mode="w") as writer:
    df.to_excel(writer, sheet_name="FixedPoint")
```
